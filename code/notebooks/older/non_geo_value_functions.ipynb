```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
grandparent_module_path = os.path.abspath(os.path.join('../..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    sys.path.append(grandparent_module_path)

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from aux_funcs import compressed_read, format_plot
```

```python
scale = 0.8
fig, axs = plt.subplots(1, 1, figsize=(4 * scale, 3 * scale))

def f(n, gamma):
    return n * np.power(gamma, n) * (1 - gamma)**2 / gamma

x = np.arange(1000)
for g in [0.95, 0.98, 0.99]:
    axs.plot(x, f(x, g), lw=1, label=r'$\gamma = $' + f'{g}')
    axs.scatter(x, f(x, g), s=3)

axs.set_xlabel(r'$k$')
axs.set_ylabel(r'$w_k$')

fig.legend()

format_plot(axs)
```

```python
scale = 0.8
fig, axs = plt.subplots(1, 1, figsize=(4 * scale, 3 * scale))

def f(n, gamma):
    return n * np.power(gamma, n) * (1 - gamma)**2 / gamma

x = np.arange(100)
for g in [0.6, 0.8, 0.95]:
    w = f(x, g)
    axs.plot(x, np.cumsum(w), lw=1, label=r'$\gamma = $' + f'{g}')
    axs.scatter(x, np.cumsum(w), s=3)

axs.set_xlabel(r'$k$')
axs.set_ylabel(r'$\sum_{i=0}^k w_{i}$')

fig.legend(loc='center right')

format_plot(axs)
```
