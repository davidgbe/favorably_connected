```python
%matplotlib inline
import sys
import os
from pathlib import Path

# Anchor to the known absolute location of this notebook
notebook_dir = Path('/home/david.bell/work/favorably_connected_2/code/notebooks/bulk_runs_jax')
code_dir = (notebook_dir / '../..').resolve()
repo_dir = (code_dir / '..').resolve()

sys.path.append(str(code_dir))
os.chdir(code_dir)

import numpy as np
import matplotlib.pyplot as plt

from aux_funcs import format_plot
from scripts.run_synthetic_agent_jax import (
    load_config_from_json,
    run_synthetic_agent,
)

plt.rcParams['font.family'] = 'Helvetica Light'
```

```python
config_path = repo_dir / 'training_configs' / 'synthetic_agent_coupled_exp.json'
```

```python
n_failures_values = [1, 2, 3, 4, 5]
```

```python
all_results = {}
for n_fail in n_failures_values:
    config = load_config_from_json(str(config_path))
    config = config.replace(
        n_failures=n_fail,
        exp_name=f'synthetic_coupled_nfail{n_fail}',
    )
    results, _ = run_synthetic_agent(config, save_trajectories=False, save_outputs=False)
    all_results[n_fail] = results['session_rewards']
```

```python
n_failures = 2
gr_scale_values = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]

gr_scale_results = {}
for gr_scale in gr_scale_values:
    config = load_config_from_json(str(config_path))
    config = config.replace(
        n_failures=n_failures,
        gr_scale=gr_scale,
        exp_name=f'synthetic_coupled_nfail{n_failures}_grscale{gr_scale:.1f}',
    )
    results, _ = run_synthetic_agent(config, save_trajectories=False, save_outputs=False)
    gr_scale_results[gr_scale] = results['session_rewards']
```

```python
hw = 0.15
fig, ax = plt.subplots(figsize=(4, 3))
for n_fail in n_failures_values:
    rewards = all_results[n_fail]
    ax.scatter([n_fail] * len(rewards), rewards,
               alpha=0.5, s=20,
               facecolors='none', edgecolors='steelblue')
    ax.plot([n_fail - hw, n_fail + hw], [np.mean(rewards)] * 2, c='black')
    print(np.mean(rewards))

ax.set_xlabel('n_failures')
ax.set_ylabel('Mean reward rate')
ax.set_xticks(n_failures_values)
ax.set_title('Synthetic agent: reward rate vs n_failures')
format_plot(ax)
fig.tight_layout()
plt.show()
```

```python
hw = 0.15
fig, ax = plt.subplots(figsize=(4, 3))
for gr_scale in gr_scale_values:
    rewards = gr_scale_results[gr_scale]
    ax.scatter([gr_scale] * len(rewards), rewards,
               alpha=0.5, s=20,
               facecolors='none', edgecolors='steelblue')
    mean_r = np.mean(rewards)
    print(mean_r)
    ax.plot([gr_scale - hw, gr_scale + hw], [mean_r, mean_r], c='black')

ax.set_xlabel('gr_scale')
ax.set_ylabel('Mean reward rate')
ax.set_xticks(gr_scale_values)
ax.set_title(f'Synthetic agent (n_failures={n_failures}):\nreward rate vs gr_scale')
format_plot(ax)
fig.tight_layout()
plt.show()
```

```python
acc_reward_scale_values = [-0.3, -0.15, 0.0, 0.15, 0.3]
n_failures_for_acc_sweep = [1, 2]
```

```python
acc_reward_scale_results = {}
for n_fail in n_failures_for_acc_sweep:
    acc_reward_scale_results[n_fail] = {}
    for acc_scale in acc_reward_scale_values:
        config = load_config_from_json(str(config_path))
        config = config.replace(
            n_failures=n_fail,
            acc_reward_scale=acc_scale,
            exp_name=f'synthetic_indep_markov_nfail{n_fail}_accscale{acc_scale:.3f}',
        )
        results, _ = run_synthetic_agent(config, save_trajectories=False, save_outputs=False)
        acc_reward_scale_results[n_fail][acc_scale] = results['session_rewards']
```

```python
colors = ['steelblue', 'darkorange']
hw = 0.008

fig, ax = plt.subplots(figsize=(5, 3))
for i, n_fail in enumerate(n_failures_for_acc_sweep):
    for acc_scale in acc_reward_scale_values:
        rewards = acc_reward_scale_results[n_fail][acc_scale]
        ax.scatter([acc_scale] * len(rewards), rewards,
                   alpha=0.4, s=15,
                   facecolors='none', edgecolors=colors[i])
        ax.plot([acc_scale - hw, acc_scale + hw], [np.mean(rewards)] * 2,
                c=colors[i], lw=2, label=f'n_fail={n_fail}' if acc_scale == acc_reward_scale_values[0] else '')

ax.set_xlabel('acc_reward_scale')
ax.set_ylabel('Mean reward rate')
ax.set_xticks(acc_reward_scale_values)
ax.set_title('Synthetic agent (indep, Markov):\nreward rate vs acc_reward_scale')
ax.legend(frameon=False)
format_plot(ax)
fig.tight_layout()
plt.show()
```
